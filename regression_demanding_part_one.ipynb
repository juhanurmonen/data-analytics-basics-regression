# Regression &ndash; challenging example one

Models for machine learning can be found in the Python's sklearn library (scikit-learn). FOr more deatils, see https://scikit-learn.org/stable/.

```python
## Import the necessary libraries.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
### In this example we use data of Boston house prices from the sklearn library.

from sklearn.datasets import load_boston
boston_data = load_boston()

### boston_data includes already named parts

boston_data.keys()
```

```python
### Description of the data.

print(boston_data.DESCR)
```

```python
### Read the data in a dataframe.

### features

df = pd.DataFrame(boston_data.data, columns = boston_data.feature_names)

### target

df['MEDV'] = boston_data.target

### Peek at the data

df.head()
```

```python
### It is always worth checking if there are missing values in the data.

df.isnull().sum()
```

```python
### distribution of the target variable (dependent variable, forecast variable)

sns.distplot(df['MEDV'], bins = 30)
```

```python
### Correlation coefficients in two decimals.

correlation_matrix = df.corr().round(2)

### Create a figure where the correlation matrix fits into.

plt.figure(figsize = (12,9))

### Coloring formatting is done by the heatmap function of the seaborn library.

sns.heatmap(data = correlation_matrix, annot = True)
```

```python
### We now choose two variables LSTAT and RM with the highest correlation coefficients with dependent variable MEDV.

features = ['LSTAT', 'RM']

target = df['MEDV']

### Add a figure big enough for two scatter charts.

plt.figure(figsize = (10,5))

### Using the function enumerate() we can look through a list and for each element return both its running number (start from zero) and value.

for i, feature in enumerate(features):
    plt.subplot(1, len(features), i+1)
    plt.scatter(df[feature], target)
    plt.xlabel(feature)
    plt.ylabel('MEDV')
```

```python
### Prepare the data for regression analysis.

### feature matrix

X = df[features]

### target variable

y = target
```

```python
### Split the data between training data and test data (20 % of the data).

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)

### Parameter random_state ensures that in each round there is the similar division between the training and test data.

### Check the sizes of emerged datas.

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
### Fitting regression model into the training data.

from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X_train, y_train)
```

Analyze results.

```python
### Coefficients

print('Coefficients: ', model.coef_)
```

```python
### Intercept

print('Intercept: ', model.intercept_)
```

```python
### Statistics related to the reliability of the model.

from sklearn.metrics import mean_absolute_error

y_train_predict = model.predict(X_train)

### Mean of absolute errors

mae = (mean_absolute_error(y_train, y_train_predict))

### R squared

r2 = model.score(X_train, y_train)

print('Suitability of the model for the training data')
print("---------------------------------------------")
print('Mean absolute error: {}'.format(mae))
print('Coefficient of determination: {}'.format(r2))
print("\n")

y_test_predict = model.predict(X_test)
mae = (mean_absolute_error(y_test, y_test_predict))
r2 = model.score(X_test, y_test)

print('Suitability of the model for the test data')
print('------------------------------------------')
print('Mean absolute error: {}'.format(mae))
print('Coefficient of determination: {}'.format(r2))
```

```python
### Compare observations of the test data and forecast of the model.

### Create a dataframe

test = pd.DataFrame()
test['y_test'] = y_test
test['y_test_predict'] = y_test_predict

### View the result as jointplot of the seaborn library.

sns.jointplot(data=test, x = 'y_test', y = 'y_test_predict', kind = 'reg')
```

Observe here that 
- Dependence between LSTAT and MEDV is rather curvilinear than linear. Here logarithmic tranformation for LSTAT variable may be considered.
- More than two variables could be used in the model.
